```python
from typing_extensions import Self
# 주문 번호 코드로 출력
print('⟝' + '-' * 30 + '⟞')
print('|' + ' ' * 31 + '|')
print('|' + ' ' * 31 + '|')
print('|' + ' ' * 31 + '|')
print('|' + ' ' * 31 + '|')
print('상품명 : HOT 아메리카노')
print('총 주문 금액 : 1500원')
print('|' + ' ' * 31 + '|')
print('|' + ' ' * 31 + '|')
print('|' + ' ' * 31 + '|')
print('|' + ' ' * 31 + '|')
print('⟝' + '-' * 30 + '⟞')

class Kiosk:
    def __init__(self):
        self.menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
        self.price = [2000, 3000, 3000, 2500, 2500, 3000]

    def menu_print(self):
        for i in range(len(self.menu)):
            print(i+1, self.menu[i], ' : ', self.price[i], '원')

    def menu_select(self):
        self.order_menu = []                                                    #추가주문 리스트
        self.order_price = []
        n = 0
        while n < 1 or len(self.menu) < n:
            n = int(input("음료의 번호를 입력하세요 : "))                       # 음료 번호 입력

            if 1 <= n & n <= len(self.menu):
                self.order_price.append(self.price[n-1])                        # 가격 리스트에 추가합니다.
                self.price_sum = self.price[n-1]

            else:                                                               # 메뉴판에 없는 번호일 때
                print("없는 메뉴입니다. 다시 주문해 주세요.")

        t = 0
        while t != 1 and t != 2:                                                # 1이나 2를 입력할 때까지 물어보기
            t= int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
            if t == 1:
                self.temp = "HOT"
            elif t == 2:
                self.temp = "ICE"
            else:
                print("1과 2 중 하나를 입력하세요.\n")

        print('주문 음료', self.temp, self.menu[n-1], ' : ', self.price[n-1], '원')

        #추가주문
        while n != 0:
            print()
            n = int(input("추가 주문은 음료 번호를, 지불은 0을 누르세요 : "))  # 추가 주문 또는 지불
            if n > 0 and n < len(self.menu) + 1:
                t = 0
                while t != 1 and t != 2:
                    t= int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
                    if t == 1:
                        self.temp = "HOT"
                    elif t == 2:
                        self.temp = "ICE"
                    else:
                        print("1과 2 중 하나를 입력하세요.\n")
                self.order_price.append(self.price[n-1])
                self.order_menu.append(self.temp + ' ' + self.menu[n-1])
                self.price_sum +=  self.price[n-1]

                print('추가 주문 음료', self.temp, self.menu[n-1], ' : ', self.price[n-1], '원\n', '합계 : ', self.price_sum, '원')
            else :
                if n == 0:
                    print("주문이 완료되었습니다")
                    print(self.order_menu, self.order_price)
                else :
                    print("없는 메뉴입니다. 다시 주문해 주세요")

    def pay(self):
        print(f'총 결제 금액은 {self.price_sum} 원 입니다')

        self.pay = 0
        while self.pay != 1 and self.pay != 2:
            self.pay = int (input ('현금 결제는 1번, 카드결제는 2번을 입력해 주세요 : '))
            if self.pay == 1:
                print('직원을 호출하겠습니다.')
            elif self.pay == 2:
                print('IC 칩 방향에 맞게 카드를 꽂아 주세요')
            else:
                print('결제를 다시 시도해 주세요')

    def table(self):
        outline1 = "⟝"  + "-" * 30 + "⟞"
        outline2 = "| "  + " " * 31 + "|"

        print(outline1)
        for i in range(5):
            print(outline2)
        for i in range(len(self.order_menu)):
            print(self.order_menu[i], ' : ', self.order_price[i])
        #print('합계 금액 :', self.prie_sum)
        print('합계 금액 :', self.price_sum)

        for i in range(5):
            print(outline2)
        print(outline1)


#Kiosk().menu
a = Kiosk()
a.menu_print()
a.menu_select()
a.pay()
a.table()


```
