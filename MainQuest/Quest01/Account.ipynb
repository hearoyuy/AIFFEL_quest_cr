```python
import random

class Account:
    accounts = []                                                               # 모든 계좌 정보를 저장하는 리스트
    account_count = 0                                                           # 생성된 계좌 갯수

    def __init__(self, name, cash):
        self.bank_name = "SC은행"
        self.name = name
        self.account_number = self.generate_account_number()
        self.cash = cash
        self.deposit_count = 0
        self.withdraw_count = 0
        self.deposit_history_list = []
        self.withdraw_history_list = []

        Account.accounts.append(self)
        Account.account_count += 1

        print(f"\n계좌가 성공적으로 개설되었습니다!")
        print(f"은행이름: {self.bank_name}, 예금주: {self.name}, 계좌번호: {self.account_number}, 잔액: {self.cash:,}원")

    @staticmethod
    def generate_account_number():
        """3자리-2자리-6자리 형태의 랜덤 계좌번호 생성"""
        return f"{random.randint(100, 999)}-{random.randint(10, 99)}-{random.randint(100000, 999999)}"

    def deposit(self, cash):
        if cash < 1:
            print("입금은 최소 1원 이상만 가능합니다.")
            return
        self.cash += cash
        self.deposit_history_list.append(cash)
        self.deposit_history_list.append((self.deposit_count + 1, cash, self.cash))
        print(f"입금횟수 :{self.deposit_count} 번")
        print(f"{cash:,}원이 입금되었습니다. 현재 잔액: {self.cash:,}원")

        if self.deposit_count % 5 == 0:
            interest = int(self.cash * 0.01)
            self.cash += interest

            print(f"{interest}원의 이자가 입금되었습니다.")

    def withdraw(self, cash):
        if cash > self.cash:
            print("계좌 잔고 이상으로 출금할 수 없습니다.")
            return
        if cash < 1:
            print("출금은 최소 1원 이상이어야 합니다.")
            return
        self.cash -= cash
        self.withdraw_history_list.append(cash)
        self.withdraw_history_list.append((len(self.withdraw_history_list) + 1, cash, self.cash))
        print(f"{cash:,}원이 출금되었습니다. 현재 잔액: {self.cash:,}원")

    def display_info(self):
        print(f"\n은행이름: {self.bank_name}, 예금주: {self.name} , 계좌번호: {self.account_number},잔액: {self.cash:} 원")

    def deposit_history(self):
        print(f"\n{self.name}님의 입금 내역:")
        for cash in self.deposit_history_list:
            print(f"+ {cash:,}원")

    def withdraw_history(self):
        print(f"\n{self.name}님의 출금 내역:")
        for cash in self.withdraw_history_list:
            print(f"- {cash:,}원")

    def find_account(account_number):
        for account in Account.accounts:
            if account.account_number == account_number:
                return account
        return None

    def display_all_accounts():
        print("\n--- 전체 계좌 정보 ---")
        for account in Account.accounts:
            account.display_info()


    def display_high_cash_accounts():
        print("\n--- 잔고 100만원 이상 고객 ---")
        for account in Account.accounts:
            if account.cash >= 1_000_000:
                account.display_info()

    def display_cashlist():
        print("\n---입출금 내역 조회 ---")



def main():
    while True:
        print("--- SC은행 ATM기 ---")
        print("1. 계좌 개설")
        print("2. 입금")
        print("3. 출금")
        print("4. 계좌 정보 보기")
        print("5. 잔고 100만원 이상 고객 보기")
        print("6. 입출금 내역보기")
        print("7. 종료")
        choice = input("원하는 작업을 선택하세요 (1-6): ")

        if choice == '1':
            name = input("예금주 이름을 입력하세요: ")
            cash = int(input("초기 입금액을 입력하세요: "))
            Account(name, cash)

        elif choice == '2':
            account_number = input("입금할 계좌번호를 입력하세요: ")
            account = Account.find_account(account_number)
            if account:
                cash = int(input("입금할 금액을 입력하세요: "))
                account.deposit(cash)
            else:
                print("계좌를 찾을 수 없습니다.")

        elif choice == '3':
            account_number = input("출금할 계좌번호를 입력하세요: ")
            account = Account.find_account(account_number)
            if account:
                cash = int(input("출금할 금액을 입력하세요: "))
                account.withdraw(cash)
            else:
                print("계좌를 찾을 수 없습니다.")

        elif choice == '4':
            Account.display_all_accounts()

        elif choice == '5':
            Account.display_high_cash_accounts()

        elif choice == '6':
            account_number = input("내역을 확인할 계좌번호를 입력하세요: ")
            account = Account.find_account(account_number)
            if account:
                account.deposit_history()
                account.withdraw_history()
            else:
                print("계좌를 찾을 수 없습니다.")

        elif choice == '7':
            print("서비스를 종료합니다.")
            break

        else:
            print("잘못된 입력입니다. 다시 선택하세요.")

main()
```
